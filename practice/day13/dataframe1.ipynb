```python
import numpy as np
import pandas as pd

file1 = pd.read_csv("../file_processing/sales_1.csv", encoding="utf-8")
file2 = pd.read_csv("../file_processing/sales_2.csv", encoding="utf-8")
file1
```

```python
file2
```

```python
# 파일 합치기
df1 = pd.concat([file1, file2])
df1
```

```python
# 인덱스 초기화 & 재정렬
df1.reset_index(drop=True, inplace=True)
df1
```

```python
# 인덱스 재조정 없이 새로운 인덱스로 dataframe 합치기
# ignore_index=True
df2 = pd.concat([file1, file2], ignore_index=True)
df2
```

```python
# 열 '판매금액'을 삽입 후, 0으로 모두 설정하세요.
df2['판매금액'] = None
df2
```

```python
# 판매금액 = 판매수량 * 판매단가로 수정
df2['판매금액'] = df2['판매수량'] * df2['판매단가']
df2
```

```python
# 응용
#1. 열 '부가세율'을 삽입 후 0으로 설정
df2['부가세율'] = 0
df2
```

```python
# 2. 부가세율을 3%인 0.03으로 재종하여 삽입
df2['부가세율'] = 0.03
df2
```

```python
# 3. 부가세액을 판매금액 * 부가세율을 곱하여 계산
df2['부가세액'] = df2['판매금액'] * df2['부가세율']
df2
```

```python
# 4. 실제판매가 = 판매금액 + 부가세액을 더하여 계산
df2['실제판매가'] = df2['판매금액'] + df2['부가세액']
df2
```

```python
df_backup = df2.copy()
df_backup
```

```python
df_sale_date = df2.groupby('판매일자').sum()
df_sale_date
```

```python
df_salepum_date = df2.groupby(['판매일자', '제품명']).sum()
df_salepum_date
```

```python
# 응용 실제 판매가에 대한 평균
sale_avg = round(np.mean(df2[['실제판매가']]),2)
print(sale_avg)
```
