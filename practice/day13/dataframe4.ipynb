```python
import numpy as np
import pandas as pd

# page 241 2번문제
df1 = pd.read_csv('../file_processing/member_1.csv')
df2 = pd.read_csv('../file_processing/member_2.csv')
df1
```

```python
df2
```

```python
# 1. id를 키값으로 하여 키값이 같은 두 데이터 프레임을 합병
df = pd.merge(df1, df2, on='id', how='inner')
df
```

```python
# 2. 열 'age'가 20세인 데이터만 추출
# df[df['age'] == 20]
df.query('age == 20')
```

```python
# 3. 열 'name'을 기준으로 오름차순 정렬하여 데이터를 표시
# HINT
# df.sort_values(by='열이름', ascending=True)
df.sort_values(by='name', ascending=True)
```

```python
# 3-1. 정렬 연습
df_sort = df.sort_values(by='id', ascending=True)
df_sort
```

```python
# 3-2. 정렬연습
df_age = df.sort_values(by=['age', 'name'], ascending=[False, True])
df_age
```

```python
# 4. 열 'age' 기준으로 나이가 많은 상위 3개의 데이터를 내림차순으로 표시
df.sort_values(by=['age'], ascending=[False]).head(3)
```

```python
# 응용
# 1. 학교 학과생들의 학과별 평균 연령은?
df.groupby('dept').agg({'age': 'mean'})
```
