```python
# page 320
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

# 1. 엑셀파일을 가져와서 판다스 데이터프레임 자료로 전환
df = pd.read_excel('../file_processing/major.xlsx')
df
```

```python
# 2. 데이터값의 결측치를 탐색
df.isnull().sum()
```

```python
# 3. 열 '글로컬영어'의 결측치를 숫자 정수 0으로 채우시오.
df.fillna({'글로컬영어': 0}, inplace=True)
df
```

```python
# 4. 열 '프로젝트명'의 결측치를 제거한 후, df 데이터 프레임에 적용하시오
df.dropna(subset='프로젝트명', inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# 5. 중복 데이터가 있는지 확인
df[df.duplicated(keep=False)]
```

```python
# 6. 중복 데이터를 제거. 제거된 인덱스에 번호 새롭게 부여
df.drop_duplicates(keep=False, ignore_index=True, inplace=True)
df
```

```python
# 7. 아래 데이터를 인덱스 15번 행 위치에 추가
df.loc[15] = ['N20', '온라인게임', '은지원', 90.0, 80, 80, 80]
df
```

```python
# 8. 열에 '총점'을 추가한 후, 총점을 구하시오.
# 총점 = '글로컬영어' + '알고리즘' + '자료구조' + 'PYTHON'
df['총점'] = df['글로컬영어'] + df['알고리즘'] + df['자료구조'] + df['PYTHON']
df
```

```python
# 9. 열에 '평균'을 추가한 후, 평균을 구하시오.
# 평균 = 총점 / 4
df['평균'] = df['총점'] / 4
df
```

```python
# 10. 글로컬영어의 실수형 자료형을 정수형 자료형으로 수정하시오.
df['글로컬영어'] = df['글로컬영어'].astype(int)
df
```

```python
# 11. 열 '전공코드' C35의 '프로젝트명'이 '바둑게임'인 '전공코드'를 'C34'로 모두 대체 하시오.
df.loc[(df['전공코드'] == 'C35') & (df['프로젝트명'] == '바둑게임'), '전공코드'] = 'C34'
df
```

```python
# 12. 열 '전공코드' 기준 오름차순으로 정렬하시오.
df.sort_values(by=['전공코드'], ascending=True)
```

```python
# 13. 열 '전공코드' 기준으로 그룹화 하시오.
df.groupby(['전공코드']).sum()
```

```python
# 14. 그룹화된 '전공코드'를 기반으로 '총점'에 대한 원형 그래프를 완성하시오.
# 단, 그래프 제목은 '학부 전공 성적 summary'로 글자색은 파랑, 글자크기는 20이다.
plt.rcParams['font.family'] = ['AppleGothic']
plt.title('학부 전공 성적 summary', color='blue', fontsize=20)
df.groupby('전공코드')['총점'].sum().plot(kind='pie')
```

```python
# 응용1. df['전공코드']가 N20이고 프로젝트명이 온라인게임인 데이터 코드를 무조건 N88로 변경
df.loc[(df['전공코드'] == 'N20') & (df['프로젝트명'] == '온라인게임'), '전공코드'] = 'N88'
df
```

```python
# 응용2. 4과목(글로컬영어, 알고리즘, 자료구조, PYTHON) 점수가 모두 80점 이상인 데이터를 조회
df[(df['글로컬영어'] >= 80) & (df['알고리즘'] >= 80) & (df['자료구조'] >= 80) & (df['PYTHON'] >= 80)]
```

```python
# 응용3. 성명이 김씨 성을 가진 사람을 조회
# HINT df['성명'].str.contains('김')
df[df['성명'].str.contains('김')]
```
