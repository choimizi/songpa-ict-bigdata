```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# page 252
plt.rcParams['font.family'] = 'nanum gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('../file_processing/traffic_accident.csv', encoding='utf-8')
df
```

```python
plt.figure(figsize=(8, 6), facecolor='gray', edgecolor='purple', linewidth=10)
data1 = df['년도']
# 사망자수가 comma 포함 및 string이어서 변환
data2 = np.array(df['사망자수(명)'].str.replace(',', '').astype(int))
plt.grid(True)
plt.plot(data1, data2, label='accident', color='b', marker='^')

plt.xlabel('연도', fontsize=18, color='blue')
plt.legend(loc='upper right', fontsize=18)
plt.ylabel('사망자수(명)', fontsize=18, color='red')
plt.title("연도별 교통사고 사망자 현황", fontsize=20)
plt.show()
```

```python
data1 = df['년도']
# 사망자수가 comma 포함 및 string이어서 변환
data2 = np.array(df['사망자수(명)'].str.replace(',', '').astype(int))
plt.grid(True)
plt.plot(data1, data2, label='accident', color='b', marker='o')

fig = plt.figure(figsize=(8, 6), facecolor='gray', edgecolor='purple', linewidth=10)
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(data2, 'ko-')

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(data2, 'r-', drawstyle='steps-post')

plt.title("연도별 교통사고 사망자 현황", fontsize=20)
plt.show()
```

```python
data1 = df['년도']
# 사망자수가 comma 포함 및 string이어서 변환
data2 = np.array(df['사망자수(명)'].str.replace(',', '').astype(int))
# 부상자수가 comma 포함 및 string이어서 변환
data3 = np.array(df['부상자수(명)'].str.replace(',', '').astype(int))

plt.grid(True)
plt.plot(data1, data2, label='accident', color='b', marker='o')

fig = plt.figure(figsize=(8, 6), facecolor='gray', edgecolor='purple', linewidth=10)
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(data2, 'ko:')

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(data3, 'r:', drawstyle='steps-post')

plt.title("연도별 교통사고 사망자, 부상자 현황", fontsize=20)
plt.show()
```
