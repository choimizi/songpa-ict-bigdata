```python
import numpy as np
import pandas as pd

# 1. 제공된 sungjuk.csv 파일을 폴더에 복사한 후, 전처리 작업을 진행하시오.
df = pd.read_csv('../file_processing/sungjuk.csv')
df_backup = df.copy()
df
```

```python
# 2. 결측치 값이 있는지 조회하시오.
df.isnull().sum()
```

```python
# 3. ‘kor’, ‘eng’, ‘math’ 세 과목 모두 결측치 값을 삭제하시오.
df.dropna(how='all', subset=['kor', 'eng', 'math'], inplace=True)
df
```

```python
# 4. 행 방향으로 2개 이상의 NULL 값이 있는 데이터를 모두 삭제하시오.
# a.k.a 정상인 데이터가 3이상
df.dropna(thresh=3, axis=0, inplace=True)
df
```

```python
# 5.‘math’ 과목에 모두 0 값으로 채워넣으시오.
# df['math'].fillna(0, inplace=True)
df.fillna({'math': 0}, inplace=True)
df
```

```python
# 6. ‘kor’,‘ ’eng’, ‘math’ 세과목 모두 정수 처리하시오.
df['kor'] = df['kor'].astype(int)
df['eng'] = df['eng'].astype(int)
df['math'] = df['math'].astype(int)
df
```

```python
# 7. 인덱스 번호를 순서대로 재설정하시오.
df.reset_index(drop=True, inplace=True)
df
```

```python
# 8. 8번 인덱스 위치에 아래 데이터를 삽입하시오.
df.loc[8] = ['이은비', 60, 90, 0]
df
```

```python
# 9. 중복 데이터가 있는지 조회하시오.
df[df.duplicated(keep=False)]
```

```python
# 10. 중복된 데이터 중 첫 번째 조회되는 중복 데이터는 유지하고 그 이후 중복되는 데이터를 삭제하시오.
df.drop_duplicates(keep='first', inplace=True)
df
```

```python
# 11. 데이터가 없는 ‘math’ 열을 삭제해주세요.
# del df['math']
df.drop(columns=['math'], axis=1, inplace=True)
df
```

```python
# 12. 두 과목 중 한 과목이라도 0점인 과목은 삭제해 주세요.
df.drop(index=df[(df.kor == 0) | df.eng == 0].index, inplace=True)
df
```

```python
# 13. ‘kor’ 과목이 60점 이하인 경우 65점으로 데이터를 상향 조정해 주세요.
df.loc[df.kor <= 60, 'kor'] = 65
df
```

```python
# 14. 순위를 부여하시오. 단, 순위는 kor+eng 점수에 대해 점수가 높은 순으로 순위를 부여하시오. Hint : df[‘열’].rank(method=’dense’, ascending=False)
# rank (method = average | min | max | first | dense)
df['rank'] = (df['eng'] + df['kor']).rank(method='dense', ascending=False).astype(int)
df
```

```python
# 15. 성적이 우수한 성적 순위 기준으로 정렬하시오.
df.sort_values(by='rank', ascending=True)
```

```python
# 응용1. 국어(kor) 과목과 영어(eng) 과목이 모두 85점인 데이터를 과감히 삭제해주세요.
df.drop(df[(df.kor == 85) & (df.eng == 85)].index, inplace=True)
df
```

```python
# 응용2. 최종 데이터 grade_report를 excel로 보내보자.
# 단, index는 생성하지 않고, sheet_name='grade'로
grade_report = df.sort_values(by='rank', ascending=True)
```

```python
grade_report.to_excel("grade_report.xlsx", index=False, sheet_name='grade')
```
