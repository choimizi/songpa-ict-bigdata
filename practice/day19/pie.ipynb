```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# page 249
plt.rcParams['font.family'] = 'nanum gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('../file_processing/supply.csv', encoding='utf-8')
df
```

```python
label = df['구분']
data = df['판매량']
plt.pie(data, labels=label, autopct='%.2f%%', explode=(0.01, 0.01, 0.01), shadow=True)
plt.title("판매량 집계", fontsize=18)
# plt.show()    # show를 하고 저장하면 파일이 안나온다..!ㅠㅠ
plt.savefig('pie.png')
```
