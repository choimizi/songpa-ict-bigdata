```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# page 249
plt.rcParams['font.family'] = 'nanum gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_csv('../file_processing/supply.csv', encoding='utf-8')
df
```

```python
plt.figure(figsize=(4, 4), facecolor='gray', edgecolor='purple', linewidth=10)
label = df['구분']
index = np.arange(len(label))
colors = ['#F5B7B1', '#F4D03F', '#73C6B6']
plt.bar(index, df['가격'], color=colors, width=0.5, alpha=0.5, linewidth=2)
plt.title("판매 가격", fontsize=20, color='green')
plt.xlabel('구분', fontsize=18, color='blue')
plt.ylabel('가격', fontsize=18, color='red')
plt.show()
```

```python
plt.figure(figsize=(4, 4), facecolor='gray', edgecolor='purple', linewidth=10)
label = df['구분']
index = np.arange(len(label))
colors = ['#F5B7B1', '#F4D03F', '#73C6B6']
plt.barh(index, df['가격'], color=colors, height=0.5, alpha=0.5, linewidth=2)
plt.title("판매 가격 수평 막대 그래프", fontsize=20, color='green')
plt.xlabel('가격', fontsize=18, color='blue')
plt.ylabel('구분', fontsize=18, color='red')
plt.yticks(index, label)
plt.show()
```
