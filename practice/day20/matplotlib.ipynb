```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 제공해드린 자료 “convience_쉼표.csv” 외부 파일을 c:\python_work 폴더에 복사한 후, 실습을 진행하세요. 단, csv 파일은 ,(콤마)로 구분되어 저장되어 있다.
df = pd.read_csv('../file_processing/convience_쉼표.csv')
df
```

```python
# 2. “사원코드”, “판매사원” index 기준으로 그룹화를 진행하시오.
df_group = df.groupby(['사원코드', '판매사원']).sum()
df_group
```

```python
# 3. 그룹화된 ‘시간수당’을 데이터로 활용하여 원형 그래프를 완성하시오.
plt.rcParams['font.family'] = 'nanum gothic'
plt.rcParams['axes.unicode_minus'] = False

label = df['판매사원'].unique()
data = df_group['시간수당']
plt.pie(data, labels=label, autopct='%.2f%%', explode=(0.01, 0.01, 0.01, 0.01), shadow=True)
plt.title("사원 코드별 수당 현황", fontsize=18)
plt.show()
```

```python
# 4. 그룹화된 “근무시간” 데이터로 활용하여 쪼개진 원형 그래프에 3D 입체 효과를 적용하시오.
data = df_group['근무시간']
plt.pie(data, labels=label, autopct='%.2f%%', explode=(0.01, 0.01, 0.01, 0.01), shadow=True)
plt.title("사원 근무시간 현황", fontsize=18)
plt.show()
```

```python
# 5. 그룹화된 “근무시간”를 데이터, “사원코드”를 x축으로 활용하여 세로막대 그래프를 완성하시오.
data = df_group['근무시간']
index = df['사원코드'].unique()

plt.figure(figsize=(6, 4), facecolor='yellow', edgecolor='blue', linewidth = 10)
color = ['green', 'blue', 'red', 'yellow']
plt.bar(index, data, color=color, width=0.5)
plt.xlabel('사원코드')
plt.ylabel('근무시간')
plt.title("사원 근무시간 현황", fontsize=18)
plt.show()
```
