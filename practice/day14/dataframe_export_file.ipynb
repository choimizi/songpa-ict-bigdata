```python
# 문제1. 송파 여행사 여름 성수기 예약일정에 대한 외부파일이다. 
# reservation_day.csv, hotelcode.csv 외부파일을 읽어서 다음 수행조건을 실행하시오

import numpy as np
import pandas as pd

reservation_day = pd.read_csv('reservation_day.csv', encoding='utf-8')
hotel_code = pd.read_csv('hotelcode.csv', encoding='utf-8')
reservation_day
```

```python
hotel_code
```

```python
# 1. 코드가 일치한 키 값을 기준으로 두개의 파일을 병합
reservation = pd.merge(reservation_day, hotel_code, on='코드', how='inner')
reservation
```

```python
# 2. 숙박요금 = 1박요금 * 숙박일수
reservation['숙박요금'] = reservation['1박요금'] * reservation['숙박일수']
reservation
```

```python
# 3. 열 ["코드","숙소명"]을 기준으로 데이터프레임 그룹을 진행
reservation.groupby(['코드','숙소명']).sum()
```

```python
# 4. 예약건수를 구하여 추가
reservation2 = reservation.groupby(['코드', '숙소명']).agg({'예약인원': 'sum', '숙박일수': 'sum', '숙박요금': 'sum', '입실일': 'count'})
reservation2.rename(columns={'입실일': '예약건수'})
```

```python
# 5. 숙박요금 기준으로 내림차순 정렬
reservation2.sort_values(by='숙박요금', ascending=False)
```

```python
# 외부 데이터 보내기
reservation.to_excel("송파성수기예약일정.xlsx", index=False, sheet_name="호텔예약", encoding='utf-8')
```
